```python
emotion_dict = \
{'neutral': 0,
 'happy': 1,
 'surprise': 2,
 'angry': 3,
 'sad': 4,
 'disqust': 5,
 'fear': 6}
```

```python
def one_hot(emotion):
    
    emo_lst = emotion.split(';')
    new_oh_list = [0 for _ in range(7)]
    
    for emo in emo_lst:
        new_oh_list[emotion_dict[emo]] = 1
        break
        
    return new_oh_list
```

```python
import os
import math
import glob
import time
import random
import torch
import glob

from PIL import Image
from torch.utils import data
from torchvision.transforms import RandomCrop

import numpy as np
```

```python
def _get_clip_max_span(df_data, midone, direction, max):
    idx = 0
    for idx in range(0, max):
        if midone+(idx*direction) < 0:
            return idx-1
        if midone+(idx*direction) >= len(df_data):
            return idx-1

    return idx
```

```python
def generate_clip_meta(entity_meta_data, midone, half_clip_size):
    max_span_left = _get_clip_max_span(entity_meta_data, midone, -1,
                                       half_clip_size+1)
    max_span_right = _get_clip_max_span(entity_meta_data, midone, 1,
                                        half_clip_size+1)
    clip_data = entity_meta_data[midone-max_span_left:midone+max_span_right+1]
    clip_data = _extend_clip_data(clip_data, max_span_left, max_span_right,
                                  half_clip_size)
    
    return clip_data
```

```python
def _extend_clip_data(clip_data, max_span_left, max_span_right, half_clip_size):
    if max_span_left < half_clip_size:
        for i in range(half_clip_size-max_span_left):
            clip_data.insert(0, clip_data[0])

    if max_span_right < half_clip_size:
        for i in range(half_clip_size-max_span_right):
            clip_data.insert(-1, clip_data[-1])

    return clip_data
```

```python
class EDATempDatasetAuxLosses(data.Dataset):
    def __init__(self, audio_root, csv_root, clip_length):
        super().__init__()

        # Data directories
        self.audio_root = audio_root
        self.csv_root = csv_root

        # Clip arguments
        self.clip_length = clip_length
        self.half_clip_length = math.floor(self.clip_length/2)

        EDA_df = pd.read_csv(f'{csv_root}/EDA_aug.csv')
        TEMP_df = pd.read_csv(f'{csv_root}/TEMP_aug.csv')
        gt_df = pd.read_csv(f'{csv_root}/gt_aug.csv')
        
        # join opernator with sciprt column
        joined_df = pd.merge(EDA_df, TEMP_df, on=['script', 'ts'], how='inner')
        
        self.joined_df = joined_df
        self.gt_df =gt_df
        
        self.entity_list = sorted(list(set(gt_df['script'])))
        
        joined_df.to_csv(f'{csv_save_path}/join_df_aug.csv', index=False)

    def __len__(self):
        return len(self.entity_list)

    def __getitem__(self, index):
        target_seg = self.entity_list[index]
        session, script, user, sequence = self.entity_list[index].split('_')
        
        target_df = self.joined_df[self.joined_df['script'] == target_seg]
        mid_index = random.randint(0, len(target_df)-1)
        start_idx = target_df.index[0]
        left_idx = mid_index-self.half_clip_length
        
        left_side_addic = 0
        right_side_addic = 0
        
        if left_idx < 0:
            left_side_addic = 0 - left_idx
            left_idx = 0
            
        right_idx = mid_index+self.half_clip_length
        
        cliped_df = target_df.iloc[left_idx:right_idx, :]
        
        
        EDA = cliped_df['EDA_value']
        TEMP = cliped_df['TEMP_value']
        
        
        right_side_addic = self.clip_length - left_side_addic - 1 - len(EDA)
        
        EDA = EDA.values.tolist()
        TEMP = TEMP.values.tolist()
        
        for _ in range(left_side_addic):
            EDA.insert(0, EDA[0])
            TEMP.insert(0, TEMP[0])
            
        for _ in range(right_side_addic):
            EDA.insert(-1, EDA[-1])
            TEMP.insert(-1, TEMP[-1])
        
        EDAt = torch.tensor(EDA)
        TEMPt = torch.tensor(TEMP)
        
        emotion = self.gt_df[self.gt_df['script'] == target_seg]['emotion_one_hot'].item()
        
        return EDAt, TEMPt, emotion
        
```
